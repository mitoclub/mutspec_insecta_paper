```python
heavy_genes = ['ND1', 'ND4', 'ND4L', 'ND5']
light_genes = ['A6', 'A8', 'CO1', 'CO2', 'CO3', 'Cytb', 'ND2', 'ND3', 'ND6']

heavy_aas = ['Asn_frac', 'Lys_frac', 'Gly_frac']
light_aas = ['Leu_frac', 'Phe_frac', 'Pro_frac']
```

```python
df = blat_aa.copy()

# Если в данных бывают варианты типа "Termites w/o workers" / "Cockroaches ..."
# этот блок сведёт их к двум группам
df['Organism_group'] = np.where(
    df['Organism'].str.contains('termite', case=False, na=False),
    'Termites',
    np.where(
        df['Organism'].str.contains('cockroach', case=False, na=False),
        'Cockroaches',
        df['Organism']
    )
)

#palette = {
#    'Termites': 'blue',
#    'Cockroaches': 'orange'
#}

palette = {
    'Cockroaches': '#744C37',         # medium brown
    'Termites': '#5DA98A'           # balanced green-teal
}
# --- Heavy genes ---
df_heavy = df[df['Gene_name'].isin(heavy_genes)].copy()
df_heavy_long = df_heavy.melt(
    id_vars=['Species_name', 'Organism_group', 'Gene_name'],
    value_vars=heavy_aas,
    var_name='Amino_acid',
    value_name='Fraction'
)

# --- Light genes ---
df_light = df[df['Gene_name'].isin(light_genes)].copy()
df_light_long = df_light.melt(
    id_vars=['Species_name', 'Organism_group', 'Gene_name'],
    value_vars=light_aas,
    var_name='Amino_acid',
    value_name='Fraction'
)

# --- Plot ---
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=False)

sns.boxplot(
    data=df_heavy_long,
    x='Amino_acid',
    y='Fraction',
    hue='Organism_group',
    order=heavy_aas,
    hue_order=['Termites', 'Cockroaches'],
    palette=palette,
    ax=axes[0]
)

axes[0].set_title('Heavy-strand genes')
axes[0].set_xlabel('Amino acid')
axes[0].set_ylabel('Fraction')
axes[0].tick_params(axis='x', rotation=45)

sns.boxplot(
    data=df_light_long,
    x='Amino_acid',
    y='Fraction',
    hue='Organism_group',
    order=light_aas,
    hue_order=['Termites', 'Cockroaches'],
    palette=palette,
    ax=axes[1]
)

axes[1].set_title('Light-strand genes')
axes[1].set_xlabel('Amino acid')
axes[1].set_ylabel('Fraction')
axes[1].tick_params(axis='x', rotation=45)

# Чтобы легенда была только одна
axes[0].legend(title='Organism')
axes[1].get_legend().remove()

plt.tight_layout()
plt.savefig('../figures/blat_aa_boxplots.pdf')
plt.show()
plt.close()
```

```python
# -----------------------------
# 1. Для light_genes: Pro / (Leu + Phe)
# -----------------------------
df_light = df[df['Gene_name'].isin(light_genes)].copy()

denom = df_light['Leu_frac'] + df_light['Phe_frac']
df_light['Pro_LeuPhe_ratio'] = np.where(
    denom > 0,
    df_light['Pro_frac'] / denom,
    np.nan
)

# -----------------------------
# 2. Для heavy_genes: Gly / (Asn + Lys)
# -----------------------------
df_heavy = df[df['Gene_name'].isin(heavy_genes)].copy()

denom = df_heavy['Asn_frac'] + df_heavy['Lys_frac']
df_heavy['Gly_AsnLys_ratio'] = np.where(
    denom > 0,
    df_heavy['Gly_frac'] / denom,
    np.nan
)

# -----------------------------
# 3. Сводка по группам
# -----------------------------
pro_leu_summary = (
    df_light.groupby('Organism_group')['Pro_LeuPhe_ratio']
    .agg(['count', 'mean', 'median', 'std', 'min', 'max'])
    .reset_index()
)

gly_asnlys_summary = (
    df_heavy.groupby('Organism_group')['Gly_AsnLys_ratio']
    .agg(['count', 'mean', 'median', 'std', 'min', 'max'])
    .reset_index()
)

print("Pro/(Leu+Phe) ratio for light genes:")
print(pro_leu_summary)
print()

print("Gly/(Asn+Lys) ratio for heavy genes:")
print(gly_asnlys_summary)
```

```python
def transform_table(df, organism):
    sp_frac = {}
    sp_list = []
    frac_type_list = []
    frac_val_list = []
    for sp in df['Species_name'].unique():
        sp_list.append([sp] * len(frac_types))
        frac_type_list.append(df[df['Species_name'] == sp].drop(columns=['Species_name', 'Organism']).mean().index)
        frac_val_list.append(df[df['Species_name'] == sp].drop(columns=['Species_name', 'Organism']).mean().values)
        sp_frac[sp] = df[df['Species_name'] == sp].drop(columns=['Species_name', 'Organism']).mean()
    
    
    df = pd.DataFrame()
    for i in range(len(sp_list)):
        df = pd.concat([df, pd.DataFrame([sp_list[i], frac_type_list[i], frac_val_list[i]]).T])
    df = df.rename(columns={0:'Species_name', 1:'Frac_type', 2:'Frac_val'})
    df['Organism'] = organism
    return df
```

```python
ter_fracs = fracs[fracs['Organism'] == 'Termites']
cock_fracs = fracs[fracs['Organism'] == 'Cockroaches']

cock_fracs = transform_table(cock_fracs, 'Cockroaches')
term_fracs = transform_table(ter_fracs, 'Termites')

fracs_merged = pd.concat([term_fracs, cock_fracs])
```

```python
fracs_merged_heavy = fracs_merged[fracs_merged['Frac_type'].isin(heavy_aas)]
```

```python
# -----------------------------
# HEAVY: Gly / (Asn + Lys)
# -----------------------------
heavy_ratio = (
    fracs_merged_heavy
    .pivot_table(
        index=['Species_name', 'Organism'],
        columns='Frac_type',
        values='Frac_val',
        aggfunc='first'   # если дубликатов нет
    )
    .reset_index()
)

heavy_ratio.columns.name = None

heavy_ratio['Gly_AsnLys_ratio'] = np.where(
    (heavy_ratio['Asn_frac'] + heavy_ratio['Lys_frac']) > 0,
    heavy_ratio['Gly_frac'] / (heavy_ratio['Asn_frac'] + heavy_ratio['Lys_frac']),
    np.nan
)

print("Heavy table with amino acid values and ratio:")
print(heavy_ratio[['Species_name', 'Organism', 'Gly_frac', 'Asn_frac', 'Lys_frac', 'Gly_AsnLys_ratio']])
print()

heavy_summary = (
    heavy_ratio
    .groupby('Organism')[['Gly_frac', 'Asn_frac', 'Lys_frac', 'Gly_AsnLys_ratio']]
    .agg(['count', 'mean', 'median', 'std', 'min', 'max'])
    .round(4)
)

print("Heavy summary by group:")
print(heavy_summary)
print()


# -----------------------------
# LIGHT: Pro / (Leu + Phe)
# -----------------------------
light_ratio = (
    fracs_merged_light
    .pivot_table(
        index=['Species_name', 'Organism'],
        columns='Frac_type',
        values='Frac_val',
        aggfunc='first'   # если дубликатов нет
    )
    .reset_index()
)

light_ratio.columns.name = None

light_ratio['Pro_LeuPhe_ratio'] = np.where(
    (light_ratio['Leu_frac'] + light_ratio['Phe_frac']) > 0,
    light_ratio['Pro_frac'] / (light_ratio['Leu_frac'] + light_ratio['Phe_frac']),
    np.nan
)

print("Light table with amino acid values and ratio:")
print(light_ratio[['Species_name', 'Organism', 'Pro_frac', 'Leu_frac', 'Phe_frac', 'Pro_LeuPhe_ratio']])
print()

light_summary = (
    light_ratio
    .groupby('Organism')[['Pro_frac', 'Leu_frac', 'Phe_frac', 'Pro_LeuPhe_ratio']]
    .agg(['count', 'mean', 'median', 'std', 'min', 'max'])
    .round(4)
)

print("Light summary by group:")
print(light_summary)
```

```python
df['Gene_name'] = df['Gene_name'].replace({'CYTB': 'Cytb'})

# оставим только нужные гены
df = df[df['Gene_name'].isin(heavy_genes + light_genes)].copy()

# -----------------------------
# 1. LIGHT: Pro / (Leu + Phe)
#    по каждому гену и группе
# -----------------------------
light_gene_table = (
    df[df['Gene_name'].isin(light_genes)]
    .groupby(['Organism', 'Gene_name'])[['Pro_frac', 'Leu_frac', 'Phe_frac']]
    .mean()
    .reset_index()
)

light_gene_table['Ratio_type'] = 'Pro/(Leu+Phe)'
light_gene_table['Gene_ratio'] = np.where(
    (light_gene_table['Leu_frac'] + light_gene_table['Phe_frac']) > 0,
    light_gene_table['Pro_frac'] / (light_gene_table['Leu_frac'] + light_gene_table['Phe_frac']),
    np.nan
)

# -----------------------------
# 2. HEAVY: Gly / (Asn + Lys)
#    по каждому гену и группе
# -----------------------------
heavy_gene_table = (
    df[df['Gene_name'].isin(heavy_genes)]
    .groupby(['Organism', 'Gene_name'])[['Gly_frac', 'Asn_frac', 'Lys_frac']]
    .mean()
    .reset_index()
)

heavy_gene_table['Ratio_type'] = 'Gly/(Asn+Lys)'
heavy_gene_table['Gene_ratio'] = np.where(
    (heavy_gene_table['Asn_frac'] + heavy_gene_table['Lys_frac']) > 0,
    heavy_gene_table['Gly_frac'] / (heavy_gene_table['Asn_frac'] + heavy_gene_table['Lys_frac']),
    np.nan
)

# -----------------------------
# 3. объединяем в одну таблицу
# -----------------------------
light_gene_table['Gene_class'] = 'light'
heavy_gene_table['Gene_class'] = 'heavy'

ratio_by_gene = pd.concat([
    light_gene_table[['Organism', 'Gene_name', 'Gene_class', 'Ratio_type',
                      'Pro_frac', 'Leu_frac', 'Phe_frac', 'Gene_ratio']],
    heavy_gene_table[['Organism', 'Gene_name', 'Gene_class', 'Ratio_type',
                      'Gly_frac', 'Asn_frac', 'Lys_frac', 'Gene_ratio']]
], ignore_index=True)

print(ratio_by_gene)
```

```python
heavy_genes = ['ND1', 'ND4', 'ND4L', 'ND5']
light_genes = ['A6', 'A8', 'CO1', 'CO2', 'CO3', 'Cytb', 'ND2', 'ND3', 'ND6']

gene_order = ['ND2', 'CO1', 'CO2', 'A8', 'A6', 'CO3', 'ND3', 'ND5', 'ND4', 'ND4L', 'ND6', 'Cytb', 'ND1']

df = blat_aa.copy()
df['Gene_name'] = df['Gene_name'].replace({'CYTB': 'Cytb'})

df['Organism_group'] = np.where(
    df['Organism'].str.contains('termite', case=False, na=False),
    'Termites',
    np.where(
        df['Organism'].str.contains('cockroach', case=False, na=False),
        'Cockroaches',
        df['Organism']
    )
)

df = df[df['Organism_group'].isin(['Termites', 'Cockroaches'])].copy()

# -----------------------------
# ratio by gene for LIGHT
# -----------------------------
light_ratio_by_gene = (
    df[df['Gene_name'].isin(light_genes)]
    .groupby(['Organism_group', 'Gene_name'])[['Pro_frac', 'Leu_frac', 'Phe_frac']]
    .mean()
    .reset_index()
)

light_ratio_by_gene['Gene_ratio'] = np.where(
    (light_ratio_by_gene['Leu_frac'] + light_ratio_by_gene['Phe_frac']) > 0,
    light_ratio_by_gene['Pro_frac'] / (light_ratio_by_gene['Leu_frac'] + light_ratio_by_gene['Phe_frac']),
    np.nan
)

# -----------------------------
# ratio by gene for HEAVY
# -----------------------------
heavy_ratio_by_gene = (
    df[df['Gene_name'].isin(heavy_genes)]
    .groupby(['Organism_group', 'Gene_name'])[['Gly_frac', 'Asn_frac', 'Lys_frac']]
    .mean()
    .reset_index()
)

heavy_ratio_by_gene['Gene_ratio'] = np.where(
    (heavy_ratio_by_gene['Asn_frac'] + heavy_ratio_by_gene['Lys_frac']) > 0,
    heavy_ratio_by_gene['Gly_frac'] / (heavy_ratio_by_gene['Asn_frac'] + heavy_ratio_by_gene['Lys_frac']),
    np.nan
)

# правильный порядок отдельно
light_order = [g for g in gene_order if g in light_genes]
heavy_order = [g for g in gene_order if g in heavy_genes]

# задаём категориальный порядок
light_ratio_by_gene['Gene_name'] = pd.Categorical(
    light_ratio_by_gene['Gene_name'],
    categories=light_order,
    ordered=True
)
light_ratio_by_gene = light_ratio_by_gene.sort_values('Gene_name')

heavy_ratio_by_gene['Gene_name'] = pd.Categorical(
    heavy_ratio_by_gene['Gene_name'],
    categories=heavy_order,
    ordered=True
)
heavy_ratio_by_gene = heavy_ratio_by_gene.sort_values('Gene_name')

palette = {
    'Termites': 'blue',
    'Cockroaches': 'orange'
}

# -----------------------------
# LIGHT plot
# -----------------------------
plt.figure(figsize=(10, 6))
ax = sns.lineplot(
    data=light_ratio_by_gene,
    x='Gene_name',
    y='Gene_ratio',
    hue='Organism_group',
    hue_order=['Termites', 'Cockroaches'],
    marker='o',
    linewidth=2,
    palette=palette
)

ax.set_title('Pro / (Leu + Phe) ratio by gene (light genes)')
ax.set_xlabel('Gene name')
ax.set_ylabel('Mean ratio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../figures/light_ratio_by_gene.pdf')
plt.show()
plt.close()

# -----------------------------
# HEAVY plot
# -----------------------------
plt.figure(figsize=(8, 6))
ax = sns.lineplot(
    data=heavy_ratio_by_gene,
    x='Gene_name',
    y='Gene_ratio',
    hue='Organism_group',
    hue_order=['Termites', 'Cockroaches'],
    marker='o',
    linewidth=2,
    palette=palette
)

ax.set_title('Gly / (Asn + Lys) ratio by gene (heavy genes)')
ax.set_xlabel('Gene name')
ax.set_ylabel('Mean ratio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../figures/heavy_ratio_by_gene.pdf')
plt.show()
plt.close()
```
